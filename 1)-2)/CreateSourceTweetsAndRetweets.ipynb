Load source-tweets

```python
path="..\\RumourTweetDatasets"
```

```python
topic="prince-toronto"
```

```python
sourceTweetsPath = path + "\\" + topic + "\\*\\source-tweets\\*"
```

```python
filePathLists = []
for file in glob.iglob(sourceTweetsPath, recursive=True):
    filePathLists.append(file)
```

```python
filePathLists
```

```python
sourceTweetsLists = []
for sourceTweetsPath in filePathLists:
    tweet = open(sourceTweetsPath)
    data = json.load(tweet)
    sourceTweetsLists.append(data)
    tweet.close()
```

```python
df = pd.json_normalize(sourceTweetsLists)
df.info()
```

```python
pd.json_normalize(sourceTweetsLists)
```

Insert source-tweets into MongoDB

```python
sourceTweetsCol.drop()
sourceTweetsCol.insert_many(sourceTweetsLists)
sourceTweetsCol.count_documents({})
```

Load retweets

```python
reTweetsPath = path + "\\" + topic + "\\*\\retweets.json"
```

```python
filePathLists = []
for file in glob.iglob(reTweetsPath, recursive=True):
    filePathLists.append(file)
```

```python
filePathLists
```

```python
reTweetsLists = []
for reTweetsPath in filePathLists:
    try:
        with open(reTweetsPath) as file:
            for line in file:
                reTweetsLists.append(json.loads(line))
            file.close()
#             print("Equipment data has been successfully retrieved.")
    except json.decoder.JSONDecodeError:
        print("There was a problem accessing the equipment data.")
```

```python
df = pd.json_normalize(reTweetsLists)
df.info()
```

```python
pd.json_normalize(reTweetsLists)
```

Insert retweets into MongoDB

```python
reTweetsCol.drop()
reTweetsCol.insert_many(reTweetsLists)
reTweetsCol.count_documents({})
```

```python
labelPath = path + "\\" + topic + "\\*\\annotation.json"
```

```python
filePathLists = []
for file in glob.iglob(labelPath, recursive=True):
    filePathLists.append(file)
```

```python
filePathLists
```
