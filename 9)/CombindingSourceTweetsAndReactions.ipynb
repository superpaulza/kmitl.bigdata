Load source-tweets

```python
path="..\\RumourTweetDatasets"
```

```python
topic="prince-toronto"
```

```python
replyingPaths = path + "\\" + topic + "\\*\\" + "reactions" + "\\*"
```

```python
replyingPaths 
```

```python
replyingFiles = []
for reFile in glob.iglob(replyingPaths, recursive=True):
    
    tweet = open(reFile)
    data = json.load(tweet)
    replyingFiles.append(data)
    tweet.close()
    
```

```python
replyingFiles
```

```python
reactions2 = db["reactions2"] 
reactions2.drop()
reactions2.insert_many(replyingFiles)
reactions2.count_documents({})
```
